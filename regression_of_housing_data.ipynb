```python
from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
```

```python
boston = pd.read_csv('./data/boston.csv')
print(boston.head())
#boston = datasets.load_boston()
#boston.keys()
x = boston.drop('MEDV', axis=1).values
y = boston['MEDV'].values
```

```python
#predicting house price from a single feature, room
x_room = x[:,5] 
print(' type x_room is ', type(x_room), 'and shape is ', x_room.shape)
x_room = x_room.reshape(-1,1)
print(' type x_room is ', type(x_room), 'and shape is ', x_room.shape)
y = y.reshape(-1,1)
print(' type y is ', type(y), 'and shape is ', y.shape)
```

```python
#plot of room versus house price (more room, higher house price)
plt.scatter(x_room, y)
plt.ylabel('Value of house/1000 ($)')
plt.xlabel('Number of rooms')
plt.show()
```

```python
#Fitting a regression model
reg = LinearRegression()
reg.fit(x_room, y)
prediction_space = np.linspace(min(x_room), max(x_room)).reshape(-1,1)
plt.scatter(x_room, y, color='blue')
plt.plot(prediction_space, reg.predict(prediction_space), color='black', linewidth=3)
plt.ylabel('Value of house/1000 ($)')
plt.xlabel('Number of rooms')
plt.show()
```

```python
#split train and test sets and fit all features
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
reg_all = LinearRegression()
reg_all.fit(x_train, y_train)
y_prediction = reg_all.predict(x_test)
print('model r square = ', round(reg_all.score(x_test, y_test),2)) # return r square
```

```python
#5 folds crosss validation
cv_scores = cross_val_score(reg_all, x,y, cv=5) # return r^2 
print(cv_scores)
print('average r^2 for 5 fold validation is ', round(np.mean(cv_scores), 2))
```
